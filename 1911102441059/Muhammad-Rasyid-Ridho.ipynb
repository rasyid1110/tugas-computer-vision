```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread('anony.jfif')
M = np.ones(img.shape, dtype='uint8') * 75
img_bright = cv2.add(img, M)
img_dark = cv2.subtract(img, M)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.set_title('Brightness')
ax1.imshow(img_bright)
ax2.set_title('Darkness')
ax2.imshow(img_dark)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread('anony.jfif')
height, width = img.shape[:2]
R = cv2.getRotationMatrix2D((width/2, height/2), 180, 1)
img_rotation = cv2.warpAffine(img, R, (width, height))
plt.imshow(img_rotation)
```
